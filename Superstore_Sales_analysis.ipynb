Data Cleaning

```python
#checking data types and null values

sales_data.info()
```

```python
# finding null values

sales_data.isnull().sum()
```

```python
#identifying rows with missing values

sales_data[sales_data['Postal Code'].isnull()]
```

```python
#assigning postal code for missing values

sales_data.loc[(sales_data['City'] == 'Burlington') & (sales_data['State'] == 'Vermont') & (sales_data['Postal Code'].isna()), 'Postal Code'] = '05401'
```

```python
sales_data.isnull().sum()
```

```python
#check for duplicate values

sales_data.duplicated().sum()
```

```python
#changing data types of required columns

sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'], format='%d/%m/%Y')
sales_data['Ship Date'] = pd.to_datetime(sales_data['Ship Date'], format='%d/%m/%Y')
```

UNIVARIATE ANALYSIS


```python
sls = sales_data['Sales'].describe()
sls
```

SALES

```python
plt.bar(sls.index, sls.values)
plt.xlabel("Statistic")
plt.ylabel("Value")
plt.title("Descriptive Statistics of Sales")
plt.xticks(rotation=45, ha='right')
plt.show()
```

SEGMENT


```python
seg = sales_data['Segment'].value_counts()
plt.figure(figsize=(8,6))
seg.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Distribution of Segments')
plt.show()
```

REGION

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.countplot(data=sales_data,x='Region',color='#00008B')
plt.title('Order count by region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

SHIPMODE

```python
plt.figure(figsize=(10,6))
sales_data['Ship Mode'].value_counts().plot(kind='bar',color='#00008B')
plt.title('Order Count by Ship Mode')
plt.xlabel('Ship Mode',)
plt.ylabel('Count')
plt.show()
```

CATEGORY


```python
cat = sales_data['Category'].value_counts()
plt.figure(figsize=(8,6))
cat.plot(kind='bar',color='#5d3e0d')
plt.title('Order count by categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

SUB-CATEGORY

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.countplot(data=sales_data,x='Sub-Category',color='#FF7F50')
plt.title('Order COunt By Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

BIVARIATE ANALYSIS





1. SALES BY REGION

```python
region_sales = sales_data.groupby('Region')['Sales'].sum().reset_index().sort_values(by='Sales',ascending=False)
region_sales
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=region_sales,x='Region',y='Sales',palette='viridis')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

2.Sales by Segment


```python
segment_sales = sales_data.groupby("Segment")['Sales'].sum().reset_index().sort_values(by='Sales',ascending=False)
segment_sales
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=segment_sales,x='Segment',y="Sales",palette='coolwarm')
plt.title('Sales by Segment')
plt.xlabel('Segemnt')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

3.Sales by Ship mode

```python
shipmode_sales = sales_data.groupby("Ship Mode")['Sales'].sum().reset_index().sort_values(by='Sales',ascending=False)
shipmode_sales

plt.figure(figsize=(10,6))
sns.barplot(data=shipmode_sales,x='Ship Mode',y="Sales",palette='magma')
plt.title('Sales by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

4.Sales by Category


```python
category_sales = sales_data.groupby("Category")['Sales'].sum().reset_index().sort_values(by='Sales',ascending=False)
category_sales

plt.figure(figsize=(10,6))
sns.barplot(data=category_sales,x='Category',y="Sales",palette='Set2')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

5.sales by Sub-Category

```python
subcat_sales = sales_data.groupby("Sub-Category")['Sales'].sum().reset_index().sort_values(by='Sales',ascending=False)
subcat_sales

plt.figure(figsize=(14,6))
subcat_sales.plot(kind='bar',x='Sub-Category',y='Sales',color='#00008B')
plt.title('Total Sales by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

6.Top 10 Sates by Sales

```python
state_sales = sales_data.groupby('State')['Sales'].sum().reset_index().sort_values(by='Sales',ascending=False).head(10)
state_sales

plt.figure(figsize=(12,6))
sns.barplot(data=state_sales,x='State',y='Sales',palette='crest')
plt.title('Top 10 States by Sales')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

7.Top 10 cities by Sales

```python
city_sales = sales_data.groupby('City')['Sales'].sum().reset_index().sort_values(by='Sales',ascending=False).head(10)
city_sales

plt.figure(figsize=(12,6))
sns.barplot(data=city_sales,x='City',y='Sales',palette='viridis')
plt.title('Top 10 Cities by Sales')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```
